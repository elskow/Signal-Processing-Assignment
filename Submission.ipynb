```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Change into wav purposes
!pip install soundfile
import soundfile as sf

# Fft transformation
from scipy.fftpack import fft, ifft
```

```python
Signal = pd.read_csv('Datasets/Data_sinyal.csv')
Noise = pd.read_csv('Datasets/Data_noise.csv')
```

```python
Signal.head()
```

# 1'st Task - Change into .wav format

```python
sf.write('output/Signal.wav', Signal['0'], 48000)
```

# 2'nd Task - Input for FFT Transformation

```python
signal_np = np.array(Signal['0'])
X = fft(signal_np)
```

# 3'rd Task - FFT plot

```python
# Plot from signal_np and result of fft

plt.figure(figsize=(20, 10))
plt.style.use('classic')
plt.subplot(2, 1, 1)
plt.plot(signal_np)
plt.title('Signal')

plt.style.use('ggplot')
plt.subplot(2, 1, 2)
plt.plot(X)
plt.title('FFT Signal')
plt.show()
```

# 4'th Task - Add noise to the signal

```python
noise_np = np.array(Noise['0'])
signal_noise_np = signal_np + noise_np
```

# 5'th Task - Noise added into input for FFT Transformation

```python
X_noise = fft(signal_noise_np)
```

# 6'th Task - FFT plot with noise

```python
plt.figure(figsize=(20, 10))
plt.style.use('classic')
plt.subplot(2, 1, 1)
plt.plot(signal_noise_np)
plt.title('Signal + Noise')

plt.style.use('ggplot')
plt.subplot(2, 1, 2)
plt.plot(X_noise)
plt.title('FFT Signal + Noise')

plt.show()
```

```python
sf.write('output/Signal_noise.wav', signal_noise_np, 48000)
```
